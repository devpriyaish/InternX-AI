```python
# Importing from modules and packages
import json
from pathlib import Path
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

print(json.dumps({"pandas": pd.__version__, "sklearn": "available"}, indent=2))
```

```python
# Custom exceptions
class DataValidationError(Exception):
    pass

def ensure_positive(numbers):
    if any(n <= 0 for n in numbers):
        raise DataValidationError("All values must be positive")
    return numbers

try:
    print(ensure_positive([1, 2, 3.1]))
except DataValidationError as exc:
    print(exc)
```

```python
# Create a tiny dataset and save it as JSON
dataset = [
    {"text": "great model performance", "label": "positive"},
    {"text": "slow inference and bugs", "label": "negative"},
    {"text": "excellent accuracy and reliability", "label": "positive"}
]
Path("mini_dataset.json").write_text(json.dumps(dataset, indent=2), encoding="utf-8")
print(Path("mini_dataset.json").read_text(encoding="utf-8"))
```

```python
# Load the dataset with pandas
df = pd.read_json("mini_dataset.json")
print(df.head())
print(df["label"].value_counts())
```

```python
# Vectorize text and prepare train/test splits
texts = df["text"].tolist()
labels = df["label"].tolist()

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(texts)
print(X.shape)
print(vectorizer.get_feature_names_out())
```

## Mini project: build a preprocessing function

Write a function that takes a raw text document and returns cleaned tokens, lowercase words, and a feature count.


```python
import re

def preprocess_text(text):
    tokens = re.findall(r"[a-zA-Z]+", text.lower())
    return tokens

sample = "AI models are awesome! They excel at language tasks."
print(preprocess_text(sample))
```
