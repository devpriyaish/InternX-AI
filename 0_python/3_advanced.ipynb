```python
# Decorators and generators
from functools import wraps

def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Running: ", func.__name__)
        return func(*args, **kwargs)
    return wrapper

@timer
def train_model(n):
    return n * 2

print(train_model(5))
```

```python
# Context managers and iterators
from contextlib import contextmanager

@contextmanager
def temporary_log(message):
    print(f"[START] {message}")
    try:
        yield
    finally:
        print(f"[END] {message}")

with temporary_log("training run"):
    print("Training loop running...")
```

```python
# Dataclasses and type hints
from dataclasses import dataclass
from typing import List

@dataclass
class ModelConfig:
    learning_rate: float
    epochs: int
    batch_size: int

config = ModelConfig(learning_rate=0.01, epochs=10, batch_size=32)
print(config)
```

```python
# Async example for AI pipelines
import asyncio

async def fetch_batch(batch_id):
    await asyncio.sleep(0.01)
    return f"batch-{batch_id}"

async def main():
    tasks = [fetch_batch(i) for i in range(3)]
    print(await asyncio.gather(*tasks))

asyncio.run(main())
```

```python
# Build a reproducible sklearn pipeline
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = np.array([[0, 1], [1, 2], [3, 4], [4, 5]], dtype=float)
y = np.array([0, 0, 1, 1])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
pipeline = Pipeline([
    ("scale", StandardScaler()),
    ("clf", LogisticRegression())
])
pipeline.fit(X_train, y_train)
print(pipeline.score(X_test, y_test))
```

## Mini project: create your own evaluation helper

Design a function that accepts predictions and true labels, then outputs accuracy and a confusion matrix summary.


```python
from collections import Counter

def evaluate_predictions(y_true, y_pred):
    accuracy = sum(int(a == b) for a, b in zip(y_true, y_pred)) / len(y_true)
    return {"accuracy": accuracy, "distribution": dict(Counter(zip(y_true, y_pred)))}

result = evaluate_predictions([1, 0, 1, 1], [1, 0, 0, 1])
print(result)
```
