### Conditionals and loops

## Mini project: simple text classifier features

Use a small list of words and count how many times each appears. This is the foundation of bag-of-words features for text models.


```python
documents = [
    "AI is fascinating",
    "Python is great for AI",
    "AI and Python work well together"
]

vocab = sorted({word for doc in documents for word in doc.lower().split()})
feature_matrix = []
for doc in documents:
    words = doc.lower().split()
    feature_matrix.append([words.count(word) for word in vocab])

print(vocab)
print(feature_matrix)
```

## Challenge
- Build a small function that cleans punctuation from a sentence.
- Create a list of 5 animal names and find the longest one.
- Save a tiny JSON file for a future lesson.


```python
import string


def clean_punctuation(sentence: str) -> str:
    """Return the sentence with punctuation removed."""
    return sentence.translate(str.maketrans("", "", string.punctuation))


example = "Hello, world! AI is great..."
cleaned = clean_punctuation(example)
print("Original:", example)
print("Cleaned :", cleaned)
```
