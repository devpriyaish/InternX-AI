## Production-ready Python habits for AI

These principles help you avoid brittle code when working with datasets, models, and APIs.

- Keep configuration in one place
- Separate data loading, preprocessing, training, and inference
- Log your experiments and version your outputs
- Prefer explicit, typed interfaces where possible


```python
from dataclasses import dataclass
from typing import List

@dataclass
class ExperimentConfig:
    model_name: str
    learning_rate: float
    epochs: int

cfg = ExperimentConfig(model_name="demo-model", learning_rate=0.01, epochs=5)
print(cfg)
```

```python
# Lightweight retrieval pipeline concept
import numpy as np

documents = [
    "Python is ideal for building AI workflows",
    "Embeddings help find semantically similar text",
    "Evaluation is necessary for reliable AI systems"
]

embeddings = np.array([
    [1.0, 0.1, 0.0],
    [0.2, 0.9, 0.1],
    [0.1, 0.2, 1.0]
])
query = np.array([0.9, 0.1, 0.0])
scores = embeddings @ query
best_idx = int(np.argmax(scores))
print(documents[best_idx])
```

```python
# Prompt structure example
system_prompt = "You are an AI tutor. Answer only with clear, practical guidance."
user_prompt = "Explain how Python helps build AI systems in 3 bullets."

conversation = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
print(conversation)
```

```python
# Simple evaluation scaffold for AI outputs
def evaluate_response(response_text, expected_keywords):
    hits = [kw for kw in expected_keywords if kw.lower() in response_text.lower()]
    return {"score": len(hits) / len(expected_keywords), "hits": hits}

sample = "Python helps with data cleaning, model training, and deployment."
print(evaluate_response(sample, ["data cleaning", "model training", "deployment"]))
```
